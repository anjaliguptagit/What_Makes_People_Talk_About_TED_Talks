```python
%matplotlib inline
import statistics
import re
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

df = pd.read_csv('ted_main.csv')
```

```python
a=list()
for index,val in df['comments'].iteritems():
        a.append(val)     
    
mean=np.mean(a)
median=np.median(a)
plt.hist(a,range=(0,500),bins=25)
plt.title("Distribution of Comments using histogram",fontsize=18)
plt.xlabel("Number of comments",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.show()
#mode=statistics.mode(a)
print("mean number of comments: ",mean)
print("median number of comments: ",median)
c=list()
for index,val in df['views'].iteritems():
        c.append(val/1000000)     
    
mean=np.mean(c)
median=np.median(c)
#mode=statistics.mode(c)
print("mean number of views: ",mean)
print("median number of views: ",median)
#print("mode number of comments: ",mode)
#plt.boxplot(df.index,df.comments)
plt.scatter(c,a)
plt.ylabel("comments")
plt.xlabel("views")
plt.show()
#df['comments'].value_counts()
```

```python
b=list()
for index,val in df['languages'].iteritems():
        b.append(val)     
    
mean=np.mean(b)
median=np.median(b)
mode=statistics.mode(b)
print("mean number of languages: ",mean)
print("median number of languages: ",median)
print("mode number of languages: ",mode)
plt.hist(df.languages)
plt.title("Distribution of Languages using Histogram",fontsize=18)
plt.xlabel("Number of languages",fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
c=list()
list_views=list()
for index,val in df['views'].iteritems():
        c.append(int(val/100000))
#list_views=pd.cut(c, bins=10)
#list_views.value_counts()
mean=np.mean(c)
median=np.median(c)
#mode=statistics.mode(c)
print("mean number of views: ",mean)
print("median number of views: ",median)
#print("mode number of views: ",mode)
#plt.boxplot(c)
#plt.scatter(df.index,c)
plt.hist(c,range=(0,50),bins=25)
plt.title("Distribution of views per lakh using histogram",fontsize=18)
plt.xlabel("Number of views per lakh",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.show()
```
